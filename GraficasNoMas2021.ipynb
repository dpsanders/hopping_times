La caja de 2D está "acostada", es decir, mide más en la dirección 1 que en la 2. Sus medidadas son
h1=1.5, h2=1.0
Por lo tanto su "vertical hop" es el que tiene el rango mas amplio.
La caja 3D esta "parada", con medidas:
h1=1, h2=1.5, h3=1.75.
Por ende su "horizontal hop" es el que tiene el rango mas amplio.
Entonces vamos a comparar Horizontal3D contra Vertical 2D.
Asi quedo, no me molesten.

```julia
datos2d=load("verthop2d-10e7-20201217.jld")
hop2d=datos2d["num_vhop_times"]
teohop2d=datos2d["exact_vhop_times"]
error2d=datos2d["vhoperror"]
rad2d=datos2d["rs"]
rvmax=rad2d[end]
rhmax=0.25
```

```julia
#datos3d=load("simulaRosa20210115-1655.jld")
datos3d=load("hoppingclosetozero3D-1e8-20210218.jld")
rad3d=datos3d["teoricas"][:,1]
hop3d=datos3d["numericas"][:,17]
error3d=datos3d["numericas"][:,19]
teohop3d=datos3d["teoricas"][:,9];
teocol3d=datos3d["teoricas"][:,2]

col3d=datos3d["numericas"][:,2]
ecol3d=datos3d["numericas"][:,3];

teowall3d=datos3d["teoricas"][:,8]
wall3d=datos3d["numericas"][:,9]
ewall3d=datos3d["numericas"][:,15];
```

```julia
plot(xlim=(0,rvmax), ylim=(0,2), xlabel="radius",ylabel="mean vertical hopping time", size=(450, 300))
scatter!(rad2d, hop2d, m=:square, ms=2, label="numerical", yerror=error2d)
plot!(rad2d, teohop2d, label="exact", lw=2)
vline!([rhmax], label="rhmax", ls=:dash)
vline!([rvmax], label="rvmax", ls=:dash)
```

```julia
plot(xlim=(0.01,rvmax), ylim=(0.5,40000000),
    xscale=:log,
    yscale=:log,
    xlabel="r_max-r",ylabel="mean vertical hopping time", size=(450, 300))
epsi=rvmax.-rad3d
scatter!(epsi, hop3d, m=:square, ms=2, label="numerical")
```

```julia
plot(xlim=(0,rvmax), ylim=(0,2), xlabel="radius",ylabel="mean vertical hopping time", size=(450, 300))
scatter!(rad2d, hop2d, m=:square, ms=2, label="2D", yerror=error2d)
scatter!(rad3d, hop3d, m=:circle, ms=3, label="3D")
vline!([rhmax], label="rhmax", ls=:dash)
vline!([rvmax], label="rvmax", ls=:dash)
```
